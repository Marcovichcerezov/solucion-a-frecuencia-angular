```
import random
import matplotlib.pyplot as plt

def moveenX():
  deltaX=random.randint(-100,100)
  return deltaX/100

def moveenY():
  deltaY=random.randint(-100,100)
  return deltaY/100

print (moveenX(), " ", moveenY())
dX = []
dY = []
Xi=0
Yi=0
X=Xi
Y=Yi
counter=0

while counter<1000:
  dX.append(X)
  dY.append(Y)
  X+=moveenX()
  Y+=moveenY()
  counter+=1
  
plt.plot(dX,dY,"r")
plt.savefig("paseo aleatorio de una particula.png")
plt.show()
```
